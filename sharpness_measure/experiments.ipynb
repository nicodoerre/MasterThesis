### Image sharpness Measure for blurred Images

```python
from utils import compute_quality_measure,convert_to_grayscale,apply_gauss_blur,load_random_image
import matplotlib.pyplot as plt
```

#### Load images

```python
dataset_path = 'C:/Users/nicol/Desktop/UNI/3.Master/MasterThesis/PracticalPart/data/DIV2K_train_HR'
img = load_random_image(dataset_path)
```

```python
img_gs = convert_to_grayscale(img)
orig_score= compute_quality_measure(img_gs)
blurred_img = apply_gauss_blur(img_gs,kernel_size=25,sigma=10)
blurred_score = compute_quality_measure(blurred_img)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(img_gs, cmap='gray')
axes[0].set_title(f"Original Image\nSharpness Score: {orig_score:.4f}")
axes[0].axis("off")

axes[1].imshow(blurred_img, cmap='gray')
axes[1].set_title(f"Blurred Image\nSharpness Score: {blurred_score:.4f}")
axes[1].axis("off")

plt.show()
```
